```python
# Merge order_items dengan products
sales_product_df = pd.merge(
    left=order_items,
    right=products,
    how="left",
    left_on="product_id",
    right_on="product_id"
)

# Merge sales_product_df dengan orders
sales_orders_df = pd.merge(
    left=sales_product_df,
    right=orders,
    how="left",
    left_on="order_id",
    right_on="order_id"
)

# Merge sales_orders_df dengan customers
sales_cust = pd.merge(
    left=sales_orders_df,
    right=customer,
    how="left",
    left_on="customer_id",
    right_on="customer_id"
)
all_df = pd.merge(
    left=sales_cust,
    right=product_category,
    how="left",
    left_on="product_category_name",
    right_on="product_category_name"
)

# Setelah merge, Anda dapat menampilkan beberapa baris pertama dari DataFrame yang dihasilkan
all_df.head()
```

4. Most review comment title

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
from collections import Counter

# Menggabungkan semua komentar menjadi satu teks
text = ' '.join(order_reviews['review_comment_title'].dropna())

# Membuat word cloud
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(text)

# Menampilkan word cloud
plt.figure(figsize=(10, 6))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()

# Menghitung frekuensi masing-masing kata
word_freq = Counter(text.split())

# Mengambil 10 kata teratas
top_10_words = word_freq.most_common(10)

# Membuat diagram batang untuk 10 kata teratas
words, frequencies = zip(*top_10_words)
plt.figure(figsize=(10, 6))
plt.bar(words, frequencies, color='skyblue')
plt.xlabel('Words')
plt.ylabel('Frequency')
plt.title('Top 10 Most Frequent Words in Reviews')
plt.xticks(rotation=45)
plt.show()
```

```python
from collections import Counter
import matplotlib.pyplot as plt

# Filter data untuk skor ulasan = 1 dan 5
score_1_comments = order_reviews[order_reviews['review_score'] == 1]['review_comment_title']
score_5_comments = order_reviews[order_reviews['review_score'] == 5]['review_comment_title']

# Fungsi untuk menghitung frekuensi kata
def count_words(comments):
    words = ' '.join(comments).split()
    return Counter(words).most_common(5)

# Mendapatkan lima kata teratas untuk skor ulasan = 1 dan 5
top_5_words_score_1 = count_words(score_1_comments)
top_5_words_score_5 = count_words(score_5_comments)

# Membuat dua subplot
fig, axs = plt.subplots(1, 2, figsize=(15, 6))

# Diagram untuk skor ulasan = 1
axs[0].bar(*zip(*top_5_words_score_1), color='salmon')
axs[0].set_title('Top 5 Words in Review Titles (Score = 1)')
axs[0].set_xlabel('Words')
axs[0].set_ylabel('Frequency')

# Diagram untuk skor ulasan = 5
axs[1].bar(*zip(*top_5_words_score_5), color='skyblue')
axs[1].set_title('Top 5 Words in Review Titles (Score = 5)')
axs[1].set_xlabel('Words')
axs[1].set_ylabel('Frequency')

# Menampilkan diagram
plt.tight_layout()
plt.show()
```
